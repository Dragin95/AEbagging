```python
import h5py 
toto = []
file_ = h5py.File("AE_train.h5", "r")
train = file_["train"]
for key in train:
    # print(f"Event:{key}")
    subdata = train[key]
    for skey in subdata:
        data = subdata[skey]
        toto.append(data)
        print(f"  |-Station{skey}-Data:{data.shape}")
        for att in data.attrs:
            print(f"  |  |--{att}:{data.attrs[att]}")
```
