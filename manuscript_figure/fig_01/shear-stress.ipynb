```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator 
import scipy.io as sio
import os
plt.rcParams['font.sans-serif']='times new roman'
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['xtick.major.width'] = 2
plt.rcParams['ytick.major.width'] = 2
plt.rcParams['xtick.major.size']  = 8
plt.rcParams['xtick.minor.size']  = 4
plt.rcParams['ytick.major.size']  = 8
plt.rcParams['ytick.minor.size']  = 4

disp=sio.loadmat('disp.mat')['disp'][:,0]
shear=sio.loadmat('shear.mat')['shear'][:,0]
time=sio.loadmat('time.mat')['time'][0,:]

# fig, ax=plt.subplots(1,1,figsize=(8,5.5),dpi=100)
# ax.plot(time, shear, 'k', linewidth=1.5, alpha=1.0)
# ax.set_xlim(3000,6000)
# ax.set_ylim(3.0,4.5)
# ax.minorticks_on()
# ax.tick_params(axis='both', labelsize=18)
# ax.set_xlabel('Elapse Time (s)',fontsize=20)
# ax.set_ylabel('Shear stress (MPa)', fontsize=20)

# ax2 = ax.twinx()
# lin2=ax2.plot(time, disp, 'r', linewidth=1.0, alpha=1.0)
# ax2.set_ylim(7,10)
# ax2.minorticks_on()
# ax2.tick_params(axis='both', labelsize=18)
# ax2.tick_params(axis='y', which='both',colors='r')
# ax2.set_ylabel('Displacement (mm)', fontsize=20)
# ax2.spines['right'].set_color('r')
# ax2.yaxis.label.set_color('r')
# y_major_locator=MultipleLocator(0.5)
# ax.yaxis.set_major_locator(y_major_locator)
# plt.tight_layout()

t1=time[0:581000]
t2=time[666000:]-850
t3=np.hstack((t1,t2))

s1=shear[0:581000]
s2=shear[666000:]
s3=np.hstack((s1,s2))

d1=disp[0:581000]
d2=disp[666000:]
d3=np.hstack((d1,d2))


fig, ax=plt.subplots(1,1,figsize=(8,5.5),dpi=100)
ax.plot(t3, s3, 'k', linewidth=1.5, alpha=1.0)
ax.set_xlim(3000,6000)
ax.set_ylim(2.5,4.5)
ax.minorticks_on()
ax.tick_params(axis='both', labelsize=20)
ax.set_xlabel('Time (s)',fontsize=22)
ax.set_ylabel('Shear stress (MPa)', fontsize=22)

ax2 = ax.twinx()
lin2=ax2.plot(t3, d3, 'r', linewidth=1.5, alpha=1.0)
ax2.set_ylim(7,10)
ax2.minorticks_on()
ax2.tick_params(axis='both', labelsize=20)
ax2.tick_params(axis='y', which='both',colors='r')
ax2.set_ylabel('Displacement (mm)', fontsize=22)
ax2.spines['right'].set_color('r')
ax2.yaxis.label.set_color('r')
y_major_locator=MultipleLocator(0.5)
ax.yaxis.set_major_locator(y_major_locator)
plt.tight_layout()
plt.savefig('shear_stress.png', dpi=300)
```
